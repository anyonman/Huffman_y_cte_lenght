```python
import numpy as np
import string
```

```python
class node:
    def __init__(self, freq, symbol, left=None, right=None):
        # frequency of symbol
        self.freq = freq
  
        # symbol name (character)
        self.symbol = symbol
  
        # node left of current node
        self.left = left
  
        # node right of current node
        self.right = right
  
        # tree direction (0/1)
        self.huff = ''
  
# utility function to print huffman
# codes for all symbols in the newly
# created Huffman tree

characs = []
codes = []  

def printNodes(node, val=''):

    
    # huffman code for current node
    newVal = val + str(node.huff)
  
    # if node is not an edge node
    # then traverse inside it
    if(node.left):
        printNodes(node.left, newVal)
    if(node.right):
        printNodes(node.right, newVal)
  
        # if node is edge node then
        # display its huffman code
    if(not node.left and not node.right):
        print(f"{node.symbol} -> {newVal}")
        characs.append(f"{node.symbol}")
        codes.append(f"{newVal}")

    return characs, codes
  
def frecuencia(texto, letra):

    file = open(texto, "r")
    texto = file.read()
    c = 0
    for k in texto:
 
        # compare each character with
        # the given letter
        if k.lower() == letra:
            c += 1
 
    # return letter count
    return c
 
 
# call function and display the letter count
#print(frecuencia('prueba.txt', 'm'))

caracteres = []

file = open("alicia.txt", "r")
texto = file.read()
for k in texto:
    caracteres.append(k)

caracteres = np.array(caracteres)

carac_unicos = np.unique(caracteres)

#print(caracteres)
#print(np.unique(caracteres))

def lista_proba(texto):
  f = []
  for k in carac_unicos:#list(string.ascii_lowercase):
    x = frecuencia(texto,k)
    f.append(np.array([k,int(x)]))
    
  return np.array(f)
    
  return np.array(f)


f = lista_proba("alicia.txt")

# characters for huffman tree
#chars = ['a', 'b', 'c', 'd', 'e', 'f']
chars = f[:,0]

# frequency of characters
#freq = [ 5, 9, 12, 13, 16, 45]
freq = f[:,1].astype(int)

# list containing unused nodes
nodes = []
  
# converting characters and frequencies
# into huffman tree nodes
for x in range(len(chars)):
    nodes.append(node(freq[x], chars[x]))
  
while len(nodes) > 1:
    # sort all the nodes in ascending order
    # based on theri frequency
    nodes = sorted(nodes, key=lambda x: x.freq)
  
    # pick 2 smallest nodes
    left = nodes[0]
    right = nodes[1]
  
    # assign directional value to these nodes
    left.huff = 0
    right.huff = 1
  
    # combine the 2 smallest nodes to create
    # new node as their parent
    newNode = node(left.freq+right.freq, left.symbol+right.symbol, left, right)
  
    # remove the 2 nodes and add their
    # parent as new node among others
    nodes.remove(left)
    nodes.remove(right)
    nodes.append(newNode)
  
# Huffman Tree is ready!
characs, codes = printNodes(nodes[0])

print(characs)
print(codes)
print(len(characs))
```

```python
def mapeo(char):
    indx = np.where(np.array(characs) == char)[0][0]
    return codes[indx]

libro_codificado = []

file = open("alicia.txt", "r")
texto = file.read()
for k in texto:
    libro_codificado.append(mapeo(k))

print(np.array(libro_codificado))
```

```python
def mapeo_inv(code):
    indx = np.where(np.array(codes) == code)[0][0]
    return characs[indx]

libro_decodificado = []

for letra_codfificada in libro_codificado:
    libro_decodificado.append(mapeo_inv(letra_codfificada))

print(np.array(libro_decodificado))
```

```python
indxs = np.where(np.array(libro_decodificado)=="\n")[0]
indxs = np.array( [0] + list(indxs) )

book_decodificado = []

for i in range(len(indxs) - 1):
    chunck = np.array(libro_decodificado)[indxs[i]+1:indxs[i+1]]
    book_decodificado.append(chunck)


print(np.array(book_decodificado))
    
np.savetxt("alicia_decodificada.txt", np.array(book_decodificado), fmt='%s')
            

```

```python
def decimalToBinary(n):
    return str(bin(n).replace("0b", "")).zfill(7)

characs_binary = []

for i, charac in enumerate(characs):
    #print(charac, decimalToBinary(i))
    characs_binary.append(decimalToBinary(i))

print(characs_binary)
```
